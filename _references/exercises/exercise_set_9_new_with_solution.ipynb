### Exercise 1.4

Write a function that 
- takes a sample of size $N$ and 
- returns the sample mean of each of the first $n$ observations, where $n=1, \cdots, N$.

### Solution

```python
# Compute sample mean at each n
def sample_means(data):
    N = len(data)
    sample_mean = np.empty(N)
    for n in range(1, N):
        sample_mean[n] = np.mean(data[:n])
    return sample_mean
```

### Exercise 1.5 

Write a function that 
- takes the sample mean of each of the first $n$ observations, where $n=1, \cdots, N$, from the sample of size $N$ and
- plots the sample mean on $y$-axis and the number of sample mean $n$ on $x$-axis.

### Solution

```python
def plot_samplemeans(sample_mean):
    fig, ax = plt.subplots()
    
    N = len(sample_mean)

    ax.plot(range(N), sample_mean, alpha=0.6, label='$\\bar{X}_n$')
    ax.plot(range(N), np.zeros(N), 'k--', lw=0.5)
    ax.legend()

    plt.show()
```

```python
import quantecon as qe
```

```python
def ll_ranksize(data):
    rank_data, size_data = qe.rank_size(data)

    fig, ax = plt.subplots()

    ax.loglog(rank_data, size_data, 'o', markersize=3.0, alpha=0.5)
    ax.set_xlabel("log rank")
    ax.set_ylabel("log size")

    plt.show()
```

### Exercise 4 (Solow-Swan Growth Model)

Recall the dynamics of the Solow-Swan Growth model

$$
    k_{t+1}
    = s \frac{F(K_t, L)}{L} + (1 - \delta) k_t
    = s f(k_t) + (1 - \delta) k_t
$$

Using the Cobb–Douglas specification $f(k) = Ak^{\alpha}$, and solving for $k^*$, we get

$$
    k^* := \left( \frac{s A}{\delta} \right)^{1/(1 - \alpha)}
$$

Take a uniform grid of 5 $\alpha$ points in $[0.1, 0.5]$ and plot per capita consumption $c^*$ at the steady state, as a function of the savings rate $s$, where $0 \leq s \leq 1$. Observe the behaviour of the consumption curve at the steady rate at different values of $\alpha$.

Assume that $A = 3.5$, and $\delta=0.2$

### Solution

Steady state consumption at savings rate $s$ is given by


$$
    c^*(s) = (1-s)f(k^*) = (1-s)A(k^*)^\alpha
$$

```python
A = 3.5
alphas = np.linspace(0.1, 0.5, 5)
delta = 0.2
s_grid = np.linspace(0, 1, 1000)
```

```python
fig, ax = plt.subplots(figsize=[11, 5])

for alpha in alphas:
    # Find the k* value
    k_star = ((s_grid * A) / delta)**(1/(1 - alpha))
    # Find the c* value
    c_star = (1 - s_grid) * A * k_star ** alpha
    ax.plot(s_grid, c_star, label=r'$\alpha=$'+str(round(alpha, 2)))

ax.set_xlabel(r'$s$')
ax.set_ylabel(r'$c^*(s)$')
ax.legend()
plt.show()
```

The rate of savings which maximizes steady state level of per capita consumption is called the [Golden Rule savings rate](https://en.wikipedia.org/wiki/Golden_Rule_savings_rate).

This shows that with the increase in $\alpha$, the Golden Rule savings rate also increases as seen from the above plot.
